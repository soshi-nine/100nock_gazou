```python
import cv2
import matplotlib.pyplot as plt

# 画像を読み込む
img = cv2.imread('gaitou.jpg')  # 画像のパスを指定

if img is not None:
    # BGRからRGBに変換する
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # matplotlibを使用して画像を表示する
    plt.imshow(img1)
    plt.axis('off')  # 軸を非表示にする
    plt.show()
else:
    print("Failed to load the image.")

```

```python
#グレーに変換
import cv2
import matplotlib.pyplot as plt

# 画像を読み込む
img = cv2.imread('gaitou.jpg')  # 画像のパスを指定

if img is not None:
    # BGRからGRAY(グレー画像)に変換する
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # matplotlibを使用して画像を表示する
    plt.imshow(img2, cmap='gray')
    plt.axis('off')  # 軸を非表示にする
    plt.show()
else:
    print("Failed to load the image.")
```

```python
#2値化（白黒の2色に画像を統一する）
img3 = cv2.imread("gaitou.jpg", 0)
#閾値設定
threshould = 128
#2値化（閾値128を超えた画像を255にする）- ret(設定した閾値)とimg(2値化後の画像)を返す
ret, img3 = cv2.threshold(img3, threshould, 255, cv2.THRESH_BINARY)
#画像表示
plt.imshow(img3, cmap="gray")
plt.axis("off")
plt.show
```

```python
#大津の2値化により自動で閾値設定
img4 = cv2.imread("gaitou.jpg", 0)
#自動で閾値設定後白黒変換
ret, img4 = cv2.threshold(img4, 0, 255, cv2.THRESH_OTSU)
#画像を表示
plt.imshow(img4, cmap="gray")
plt.axis("off")
plt.show()
```

```python
#エッジ検出（輪郭をより強調させる）
import cv2
import matplotlib.pyplot as plt

#画像読み込み
img5 = cv2.imread("gaitou.jpg", 0)
#エッジ処理(エッジ処理する画像, エッジの下限閾値, エッジの上限閾値)
img_canny = cv2.Canny(img5, 50, 110)
#表示
plt.imshow(img_canny, "gray")
plt.axis("off")
plt.show()
```
